```python
# Выделение из массива только данных о ценах и преобразование типа
data = ndata[:,1:31].astype(float)
data
```

```python
type(data[1,1])
data[1,1]
```

```python
data_centr = data - np.mean(data , axis = 0)   #Центрирование данных
data_norm = data_centr/np.std(data_centr , axis = 0)   #Нормирование данных
data_norm # Проверка данных
```

```python
covarm = np.cov(data_norm , rowvar = False) 
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(covarm) #Находим собственные значения и векторы матрицы ковариации
max_index = np.argmax(eigen_values)
np.set_printoptions(precision=3)
print(eigen_values)
sorted_index = np.argsort(eigen_values)[::-1] #Сортируем по убыванию собственные значения
sorted_index
```

```python
sorted_eigenvectors = eigen_vectors[:,sorted_index]  # Выстраиваем собственные векторы соответствующим образом
n = 10                                               #Требуемое количество компонент
eigenvector_subset = sorted_eigenvectors[:,0:n]   #Выбираем собственные векторы в количестве n
```

```python
eigenvector_subset
```

```python
data_compact = np.dot(eigenvector_subset.transpose() , data_norm.transpose()).transpose()   #Компактная форма
print(data_compact)
```

```python
data_recovered = np.dot(data_compact, eigenvector_subset.transpose())   #Восстановление данных
data_recovered
```

```python
sorted_eigenvalue = sorted(eigen_values)[::-1]  # Сортируем собственные значения по убыванию
sorted_eigenvalue
```

```python
# Так как дисперсию характеризуют собственные значения. Можно считать что 100% дисперсии это сумма всех собственных значений.
# Нужно определить сколько компонент из числа отсортированных по убыванию нужно сложить чтобы получить 90%

i = 0
while (sorted_eigenvalue[i] + sum(sorted_eigenvalue[:i])) < 0.9*sum(sorted_eigenvalue):
    i +=1
print('количество компонент должно быть равно',i+1)
```
